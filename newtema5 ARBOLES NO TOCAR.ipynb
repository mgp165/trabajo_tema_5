```
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error, roc_auc_score, top_k_accuracy_score, accuracy_score, roc_curve, auc, RocCurveDisplay
import matplotlib.pyplot as plt
import random as rd
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_graphviz
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier, GradientBoostingClassifier 
import multiprocessing
import pickle
import shap
import time
from datetime import timedelta
from sklearn import tree
import graphviz
import dtreeviz
from sklearn.datasets import make_classification
import plotly.graph_objects as go
```

```
X_train = pd.read_parquet('X_train.parquet')
y_train = pd.read_parquet('y_train.parquet')

X_validation = pd.read_parquet('X_validation.parquet')
y_validation = pd.read_parquet('y_validation.parquet')

X_test = pd.read_parquet('X_test.parquet')
y_test = pd.read_parquet('y_test.parquet')
```

## cart

```
param_grid = {
    'max_depth':[2, 5,10, 15, 25],
    'max_features': [2, 3, 5, 10, None],
    'max_leaf_nodes':[4, 8, 16, None],
}

tic = time.perf_counter()

def gini(y,y_hat):
    return 2*roc_auc_score(y, y_hat)-1

metrics = []
for max_depth in param_grid['max_depth']:
    for max_features in param_grid['max_features']:
        for max_leaf_nodes in param_grid['max_leaf_nodes']:
        
            start_time = time.perf_counter()
            tree = DecisionTreeClassifier(max_depth = max_depth, max_features=max_features, max_leaf_nodes=max_leaf_nodes)
            tree.fit(X_train,y_train)

            y_train_hat_tree = tree.predict_proba(X_train)[:, 1]
            y_validation_hat_tree = tree.predict_proba(X_validation)[:, 1]

            metrics.append([
                max_depth,
                max_features,
                max_leaf_nodes,
                gini(y_train, y_train_hat_tree),
                gini(y_validation, y_validation_hat_tree),
                timedelta(seconds=time.perf_counter() - start_time),
            ])

tac = time.perf_counter()
print(f"Elapsed time: {timedelta(seconds=tac-tic)}")
```

```
metrics_tree = pd.DataFrame(
    metrics, 
    columns =['max_depth', 'max_features','max_leaf_nodes', 'Gini_train','Gini_test', 'time']
)
metrics_tree['delta'] = (metrics_tree.Gini_test-metrics_tree.Gini_train) / metrics_tree.Gini_train
metrics_tree.sort_values('delta', ascending=False, inplace=True)
round(metrics_tree,5)
```

```
tree_55 = DecisionTreeClassifier(max_depth=10, max_features=10, max_leaf_nodes=None)
tree_55.fit(X_train,y_train)

y_train_hat_tree_55 = tree_55.predict_proba(X_train)[:, 1]
y_validation_hat_tree_55 = tree_55.predict_proba(X_validation)[:, 1]
```

```
predictions55 = pd.DataFrame({"y": y_validation['booking_status_Canceled'].tolist(),"y_hat": y_validation_hat_tree_55})
predictions55['TopPercentile'] = np.ceil(100*(-predictions55.y_hat).rank(method='first',pct = True)).round(2).astype(int)

report55 = (
    predictions55
    .groupby('TopPercentile')
    .agg(
        ScoreAvg = pd.NamedAgg('y_hat',np.mean),
        TotalCount = pd.NamedAgg('y','count'),
        Class1Count = pd.NamedAgg('y',np.sum)
    )
    .reset_index()
    .sort_values('TopPercentile')
)
report55['Class0Count'] = report55.TotalCount - report55.Class1Count

report55['TotalCountCumulated'] = report55.TotalCount.cumsum()
report55['Class1CountCumulated'] = report55.Class1Count.cumsum()
report55['Class0CountCumulated'] = report55.Class0Count.cumsum()

report55['PrecisionCumulated'] = report55.Class1CountCumulated / report55.TotalCountCumulated
report55['TruePositiveRate'] = report55.Class1CountCumulated / report55.Class1Count.sum()
report55['Sensitivity'] = report55.TruePositiveRate
report55['Recall'] = report55.TruePositiveRate

report55['FalsePositiveRate'] = report55.Class0CountCumulated / report55.Class0Count.sum()
report55['Specificity'] = 1-report55.FalsePositiveRate
report55['1-Specificity'] = report55.FalsePositiveRate

report55['Class1TotalRatio'] = report55.Class1Count.sum() / report55.TotalCount.sum()
report55['Class1TrivialUplift'] = report55.Class1TotalRatio / report55.Class1TotalRatio
report55['Class1ModelUplift'] = report55.PrecisionCumulated / report55.Class1TotalRatio

report55
```

```
fig55 = go.Figure()
fig55.add_trace(
    go.Bar(
        name="Fitted model",
        x=report55.TopPercentile,
        y=report55.Class1ModelUplift
    )
)

fig55.add_trace(
    go.Scatter(
        name="Trivial model",
        x = report55.TopPercentile,
        y = report55.Class1TrivialUplift,
        hoverinfo='none',
        line = dict(color = 'red', width = 2, dash='dash')
    )
)

fig55.update_layout(
    title = 'Uplift curve',
    xaxis_title="Top Percentile", 
    yaxis_title="Model uplift",
    showlegend=True
)

fig55.show()
```
