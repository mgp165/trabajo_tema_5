```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random as rd

data = pd.read_csv('hotel_reservations.csv')
data.head(5)
```

## Depurar datos

```
newdata = data
```

```
newdata = newdata.drop(['Booking_ID', 'arrival_year', 'arrival_date'], axis=1)
```

```
newdata = newdata.drop(newdata[newdata['market_segment_type']=='Aviation'].index)
```

```
newdata = newdata.drop(newdata[newdata['market_segment_type']=='Complementary'].index)
```

```
newdata = newdata.drop(newdata[newdata['type_of_meal_plan']=='Meal Plan 3'].index)
```

```
newdata = newdata.replace({'Meal Plan 1': 1, 'Meal Plan 2': 1, 'Not Selected': 0})
```

```
newdata = newdata.drop(newdata[newdata['room_type_reserved']=='Room_Type 3'].index)
```

```
newdata = newdata.drop(newdata[newdata['avg_price_per_room']==540].index)
```

```
newdata = newdata.drop(newdata[newdata['avg_price_per_room']<=10].index)
```

```
newdata_enc = pd.get_dummies(newdata, drop_first=False, dtype='uint8')
```

```
newdata_enc
```

## data split

```
data_train = newdata_enc.sample(frac=0.7)
```

```
train_index = data_train.index
```

```
data_no_train = newdata_enc.drop(train_index, axis=0)
```

```
data_validation = data_no_train.sample(frac=0.5)
```

```
validation_index = data_validation.index
```

```
data_test = data_no_train.drop(validation_index, axis=0)
```

```
y_train = data_train[['booking_status_Canceled']]
X_train = data_train.drop(['booking_status_Not_Canceled', 'booking_status_Canceled'], axis=1)

y_validation = data_validation[['booking_status_Canceled']]
X_validation = data_validation.drop(['booking_status_Not_Canceled', 'booking_status_Canceled'], axis=1)

y_test = data_test[['booking_status_Canceled']]
X_test = data_test.drop(['booking_status_Not_Canceled', 'booking_status_Canceled'], axis=1)
```

```
y_train.to_parquet('y_train.parquet')
X_train.to_parquet('X_train.parquet')

y_validation.to_parquet('y_validation.parquet')
X_validation.to_parquet('X_validation.parquet')

y_test.to_parquet('y_test.parquet')
X_test.to_parquet('X_test.parquet')
```
